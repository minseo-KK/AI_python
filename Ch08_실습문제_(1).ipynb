```python
# 산점도 : 두 변수 간의 관계를 점으로 나타냄
height = [160, 165, 170, 175, 180]
weight = [55, 60, 68, 72, 78]

plt.scatter(height, weight) # plot 대신 scatter 사용
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
# bar plot: 카테고리별 데이터 크기를 비교
fruits = ['Apple', 'Banana', 'Orange']
sales = [100, 150, 80]

plt.bar(fruits, sales)
plt.title('Fruit Sales')
plt.xlabel('Fruit')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
#히스토그램: 특정 구간에 데이터가 얼마나 분포하는지(빈도) 보여줄 때 사용합니다.
scores = [65, 72, 75, 78, 80, 82, 85, 88, 90, 91, 93, 95, 98, 100]

plt.hist(scores, bins=5) # 데이터, 구간(bin) 개수 설정
plt.title('Exam Score Distribution')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2006, 2009, 2012, 2015, 2018]
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]
jpn = [523, 529, 536, 532, 527]

plt.figure(figsize=(5,3))# 그래프가 가로와 세로 크기를 각각 5인치와 3인치 크기 설정
plt.plot (year,kor,'r.:',label='Korea')# year 데이터를 x축으로, kor 데이터를 y축으로 하여 빨간색 점선으로 선 그래프를 그리고 범례에 'Korea'라고 표시
plt.plot(year, usa, 'b.:',label='USA' ) #범례에 'Korea'라고 표시
plt.plot(year, jpn, 'g.-', label='Japan')
plt.ylim(400, 600)  # y축 범위 설정
plt.legend() # 그래프에 범례를 추가
plt.title('Mathematics performance')# 그래프 제목을 'Mathematics performance'로 설정합니다.
plt.xlabel('year')
plt.ylabel('score')
plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2000, 2005, 2010, 2015, 2020]
single_person = [15.5, 20.0, 23.9, 27.2, 31.7]
baby_born = [640089, 438707, 470171, 438420, 272337]

plt.figure(figsize=(5,3))
 # 1인 가구 비율(single_person)과 출생아 수(baby_born) 사이의 관계를 나타내는 산점도를 그립니다.
for i in range(len(year)):
  plt.annotate(year[i],xy=(single_person[i], baby_born[i]),xytext=(5,-10),textcoords='offset points') #각 연도에 해당하는 데이터 지점 옆에 해당 연도를 표시하는 주석을 추가
plt.xlim(14,35)
plt.ylim(230000,700000)
plt.xlabel('single-person household ratio')
plt.ylabel('the number of babies born')
plt.show()
```

```python
import matplotlib.pyplot as plt

exercise ={'climbing':0,'soccer':0,'swimming':0,'tennis':0}

print('좋아하는 운동을 입력하세요.')
print('복수 개의 운동을 입력하려면 콤마로 구분하세요.')

while True:
  s = input('climbing,soccer,swimming,tennis,종료: ')
  if s=='종료':
    break
  for a in s.split(','):
    if a in exercise:
      # 해당 운동의 카운트를 1 증가시키는 연산
    else:
      print(f'{a}는 등록되지 않은 운동입니다.')

plt.figure(figsize=(5.5,3))
plt.bar(exercise.keys(), exercise.values(), color='g', width=0.6)
#plt.xticks(rotation='vertical')
plt.xlabel('exercise')
plt.ylabel('frequency')
plt.show()
```

```python
# 8.연습문제
months= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall= [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize=(12,6))     #그래프 꾸미기-그래프 크기 조종 (plot 그리기 전에 설정)
plt.plot(months, rainfall,color='b',marker='x',linestyle='--',label='rainfall')  #'bx--' (3개 다 쓰는 경우 color-marker-linestlye 순)
plt.legend()

plt.title('Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')

for i in range(len(months)):
  plt.annotate(rainfall[i], xy=(months[i], rainfall[i]), xytext=(-5,5), textcoords='offset points')

plt.show()
```
